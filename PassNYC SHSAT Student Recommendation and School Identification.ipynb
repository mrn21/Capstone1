## Preparing the Data

```python
def stripsoff(x):
    return float(x.strip('%'))

schools["School Income Estimate"] = schools["School Income Estimate"].replace('[\$,]', '', regex=True).astype(float)
schools["Student Attendance Rate"] = schools["Student Attendance Rate"].astype(str).apply(stripsoff)

schools["Percent Asian"] = schools["Percent Asian"].astype(str).apply(stripsoff)
schools["Percent Black"] = schools["Percent Black"].astype(str).apply(stripsoff)
schools["Percent Hispanic"] = schools["Percent Hispanic"].astype(str).apply(stripsoff)
schools["Percent White"] = schools["Percent White"].astype(str).apply(stripsoff)
schools["Percent Others"] = (schools["Percent Black"] + schools["Percent Hispanic"] + schools["Percent White"] + schools["Percent Asian"]).sub(100).mul(-1)
schools["Rigorous Instruction %"] = schools["Rigorous Instruction %"].astype(str).apply(stripsoff)
schools["Collaborative Teachers %"] = schools["Collaborative Teachers %"].astype(str).apply(stripsoff)
schools["Supportive Environment %"] = schools["Supportive Environment %"].astype(str).apply(stripsoff)
schools["Effective School Leadership %"] = schools["Effective School Leadership %"].astype(str).apply(stripsoff)
schools["Strong Family-Community Ties %"] = schools["Strong Family-Community Ties %"].astype(str).apply(stripsoff)
schools["Trust %"] = schools["Trust %"].astype(str).apply(stripsoff)
schools["Student Attendance Rate"] = schools["Student Attendance Rate"].astype(str).apply(stripsoff)
schools["Percent of Students Chronically Absent"] = schools["Percent of Students Chronically Absent"].astype(str).apply(stripsoff)


schools["Economic Need Index"] = schools["Economic Need Index"].fillna(schools["Economic Need Index"].mean())
schools["School Income Estimate"] = schools["School Income Estimate"].fillna(schools["School Income Estimate"].mean())
#schools["Student Attendance Rate"] = schools["Student Attendance Rate"].fillna(schools["Student Attendance Rate"].mean())
schools.head(3)
```
